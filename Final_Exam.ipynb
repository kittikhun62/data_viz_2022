#นำเข้าข้อมูล การตรวจโรค(ov)

```python
alldb =  pd.read_csv(os.path.join(path,'alldb.csv'))
alldb
```

##เลือกเฉพาะ Column ที่ต้องการ

```python
alldb = alldb[['user','RowKey','date','ov','phone','site','time','urine','PartitionKey','title','age']]
alldb
```

```python
alldb.isnull().sum()
```

## เปลี่ยน type ของข้อมูล

###user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]
alldb
```

```python
alldb.isnull().sum()
```

###site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['site'].astype(str).str.isdigit()) & (alldb['site'].str.len() == 5)] 
alldb
```

```python
alldb.isnull().sum()
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['PartitionKey'].astype(str).str.isdigit()) & (alldb['PartitionKey'].str.len() == 5)] 
alldb
```

```python
alldb.isnull().sum()
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
set(alldb['ov'])
```

```python
alldb[alldb['ov']=='ติดเชื้อ +2 ']
```

```python
alldb = alldb.drop(16931) #ลบแถว 16931 จากคอลัมน์ ov ที่มีค่า nan = 210 และ ค่า 'ติดเชื้อ +2 '
alldb
```

```python
alldb = alldb.dropna(subset=['ov']) 
alldb
```

```python
set(alldb['ov'])
```

```python
alldb.isnull().sum()
```

###date เป็น YYYY-MM-DD

```python
alldb['date'] = pd.to_datetime(alldb['date'],format="%Y-%m-%d")
alldb
```

```python
alldb['date'] = alldb[['date']].fillna(method="ffill") #เติมค่าว่างของวันที่ด้วยค่าก่อนหน้าหรือค่าข้างบน
alldb
```

```python
alldb.dtypes
```

```python
alldb.isnull().sum()
```

###time เป็น hh-mm-ss

```python
alldb['time'] = alldb[['time']].fillna(method="ffill")  #เติมค่าว่างของเวลาด้วยค่าก่อนหน้าหรือค่าข้างบน
alldb
```

```python
alldb['time'] = pd.to_datetime(alldb['time'],format='%H:%M%:%S',errors='ignore') 
alldb
```

```python
alldb.isnull().sum()
```

###urine มีค่า String

```python
alldb['urine'] = alldb[['urine']].fillna(method="ffill")  #เติมค่าว่างของ urine ด้วยค่าก่อนหน้าหรือค่าข้างบน
alldb
```
