#นำเข้าข้อมูล การตรวจโรค(ov)

```python
data_file_path = os.path.join(path,'alldb.xlsx')
alldb = pd.read_excel(data_file_path)
```

##เลือกเฉพาะ Column ที่ต้องการ

```python
alldb_new = alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']]
alldb_new
```

###ใช้ .dropna ตัดแถวข้อมูลที่มีค่า missing

```python
alldb_drop =alldb_new.dropna()
alldb_drop
```

```python
alldb_drop.dtypes  
```

## เปลี่ยน types ของข้อมูล

###user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_drop = alldb_drop[(alldb_drop['user'].astype(str).str.isdigit()) & (alldb_drop['user'].str.len() == 5)]
alldb_drop
```

###RowKey เป็น String

```python
alldb_drop['RowKey'].astype({'RowKey':'string'}).dtypes
```

###date เป็น YYYY-MM-DD

```python
import datetime #ทำการ import package datetime 
alldb_drop = pd.DataFrame(alldb_drop)
alldb_drop['date'] = pd.to_datetime(alldb_drop['date'])
alldb_drop
```

###phone เป็น String

```python
alldb_drop['phone'].astype({'phone':'string'}).dtypes
```

###site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['site'].astype(str).str.isdigit()) & (alldb['site'].str.len() == 5)]
alldb
```

###time เป็น hh-mm-ss

```python
import datetime #ทำการ import package datetime 
alldb_drop = pd.DataFrame(alldb_drop) 
alldb_drop['time'] = pd.to_datetime(alldb_drop['time'],format= '%H:%M:%S' ).dt.time
alldb_drop
```

###urine มีค่า String

```python
alldb_drop['urine'].astype({'urine':'string'}).dtypes
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
alldb_drop = alldb.dropna(subset=['ov'])
alldb_drop
```

```python
set(alldb_drop['ov'])
```

```python
alldb_drop['ov'].replace({'ปกติ (เหลืองใส)':'0',
 'ขุ่น':'0',
 'ติดเชื้อ +2 ':'0',
 'ปกติ (เหลืองใส)':'0',
 'มีตะกอน':'0',
 'มีเม็ดเลือดแดง':'0',
 'สีน้ำนม':'0',
 'สีฟ้า หรือน้ำเงิน':'0',
 'ใสเป็นน้ำเปล่า':'0'}, inplace=True)
```

```python
set(alldb_drop['ov'])
```

```python
alldb_drop['ov'] = alldb_drop['ov'].astype('int')
alldb_drop
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_drop = alldb_drop[(alldb_drop['PartitionKey'].astype(str).str.isdigit()) & (alldb_drop['PartitionKey'].str.len() == 5)]
alldb_drop
```

```python
alldb_drop.dtypes
```

เเก้ type

```python
alldb_drop['user'] = pd.to_numeric(alldb_drop['user'], errors='coerce')     
alldb_drop['date'] = pd.to_datetime(alldb_drop['date'],format="%d/%m/%Y")
alldb_drop['time'] = pd.to_datetime(alldb_drop['time'],format='%H:%M%:%S',errors='ignore')
alldb_drop['phone']=alldb_drop['phone'].astype('string')
alldb_drop['urine']=alldb_drop['urine'].astype('string')
alldb_drop['site']=alldb_drop['site'].astype('int64')
alldb_drop['PartitionKey']=alldb_drop['PartitionKey'].astype('int64')
alldb_drop['RowKey']=alldb_drop['RowKey'].astype('string')
alldb_drop.dtypes
```
